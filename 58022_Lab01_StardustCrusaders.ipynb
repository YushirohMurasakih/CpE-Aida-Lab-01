```python
#here is the function for calculating the semestral grades based on the imported file
def get_grades(df):
  prelim = df["prelim_grades"]
  midterm = df["midterm_grades"]
  finals = df["final_grades"]
  for string in df["name"]:
    semGrades = prelim * .3 + midterm * .3 + finals * .4
    df["semestral_grades"] = semGrades
  return df
  
print(get_grades(df))
```

```python
def get_class_stats(df):

  prelim = df["prelim_grades"]
  midterm = df["midterm_grades"]
  finals = df["final_grades"]
  semGrades = df["semestral_grades"]

  #Lowest
  prelimsMin = prelim.min()
  midtermsMin = midterm.min()
  finalsMin = finals.min()
  semMin = semGrades.min()

  #Highest
  prelimsMax = prelim.max()
  midtermsMax = midterm.max()
  finalsMax = finals.max()
  semMax = semGrades.max()

  # mean
  prelimsMean = prelim.mean()
  midtermsMean = midterm.mean()
  finalsMean = finals.mean()
  semMean = semGrades.mean()

  # median
  prelimsMed = prelim.median()
  midtermsMed = midterm.median()
  finalsMed = finals.median()
  semMed = semGrades.median()

  # mode
  prelimsMode = prelim.mode()
  midtermsMode = midterm.mode()
  finalsMode = finals.mode()
  semMode = semGrades.mode()

  # standard deviation
  prelimsSD = prelim.std()
  midtermsSD = midterm.std()
  finalsSD = finals.std
  semSD = semGrades.std()


  dictionary01 = {'':["Prelim Grades","Midterm Grades","Final Grades","Semestral Grades"],
        'Lowest':[prelimsMin,midtermsMin,finalsMin,semMin],
        'Highest':[prelimsMax,midtermsMax,finalsMax,semMax],
        'Mean':[prelimsMean,midtermsMean,finalsMean,semMean],
        'Median':[prelimsMed,midtermsMed,finalsMed,semMed],
        'Mode':[prelimsMode,midtermsMode,finalsMode,semMode],
        'Standard_Deviation':[prelimsSD,midtermsSD,finalsSD,semSD]
  }
  out = pd.DataFrame(dictionary01)
  return out
```
